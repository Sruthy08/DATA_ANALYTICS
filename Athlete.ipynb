```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Step:1 create dataset
X = [
    [170,65],
    [160,50],
    [180,80],
    [150,45],
    [155,48],
    [165,55]
]
Y = ['Athlete','Athlete','Athlete','Non-Athlete','Non-Athlete','Non-Athlete']
```

```python
# Step:2 Test point
test_point = [[167,58]]
```

```python
# Step:3 Train K-NN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,Y)
```

```python
# Step:4 Predict the class
prediction = knn.predict(test_point)
print("Prediction Class:",prediction[0])
```

```python
# Step:5 Visualization
# Separate classes for plotting
athlete_data = [X[i] for i in range(len(X)) if Y[i] == 'Athlete']
non_athlete_data = [X[i] for i in range(len(X)) if Y[i] == 'Non-Athlete']

# Plotting
```

```python
# convert to X Y for plotting
ath_X = [i[0] for i in athlete_data]
ath_Y = [i[1] for i in athlete_data]
non_X = [i[0] for i in non_athlete_data]
non_Y = [i[1] for i in non_athlete_data]
```

```python
# Plotting
plt.scatter(ath_X,ath_Y,color='red',label='Athlete')
plt.scatter(non_X,non_Y,color='blue',label='Non-Athlete')
plt.scatter(test_point[0][0],test_point[0][1],color='green',label='Test Point')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('K-NN Classification')
plt.legend()
plt.grid(True)
plt.show()
```
