EXPLORATORY DATA ANALYSIS(EDA)


```python
sns.heatmap(titanic.corr(numeric_only=True),annot=True,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(x=titanic['age'],y=titanic['survived'])
```

```python
sns.scatterplot(x='age',y='survived',data=titanic,hue='sex')
```

```python
sns.stripplot(x='age',y='survived',data=titanic,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='survived',y='age',data=titanic,hue='sex',dodge=True)
```

```python
plt.hist(titanic['age'],bins=40)
plt.title("HISTOGRAM")
plt.xlabel("AGE")
plt.legend()
plt.show()
```

```python
plt.hist(titanic['age'].dropna(),bins=50,color='skyblue',edgecolor='black',label="Age")
plt.title("Histogram of passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(titanic['sex'])
```

```python
titanic['sex'].value_counts().plot.pie(autopct='%1.1f%%',colors=['lightcoral','lightgreen'])
plt.title("survival Distribution")
plt.ylabel('')
plt.show()
```

```python
sns.boxplot(x=titanic['age'],color='skyblue')
plt.xlabel("age")
plt.title("Box Plot of Passenger Ages")
plt.show()
```

```python
sns.boxplot(x='survived',y='age',data=titanic,palette='Set2')
plt.title("Age D")
```
